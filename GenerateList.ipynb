```python
import numpy as np

def generate_test_scores(num_scores=100, mean=75, std_dev=15):
    """
    Generate a list of normally distributed test scores between 0 and 100.
    
    Parameters:
    num_scores (int): Number of scores to generate (default: 100)
    mean (float): Mean score of the distribution (default: 75)
    std_dev (float): Standard deviation of the distribution (default: 15)
    
    Returns:
    list: List of test scores rounded to 1 decimal place
    """
    # Generate normally distributed scores
    scores = np.random.normal(mean, std_dev, num_scores)
    
    # Clip scores to ensure they're between 0 and 100
    scores = np.clip(scores, 0, 100)
    
    # Round to 1 decimal place and convert to list
    scores = [round(score, 1) for score in scores]
    
    return scores

# Example usage
if __name__ == "__main__":
    # Generate 100 test scores
    test_scores = generate_test_scores()
    
    # Print the results
    print("Generated Test Scores:")
    print(test_scores)
    
    # Print some basic statistics
    print(f"\nNumber of scores: {len(test_scores)}")
    print(f"Average score: {sum(test_scores)/len(test_scores):.1f}")
    print(f"Minimum score: {min(test_scores)}")
    print(f"Maximum score: {max(test_scores)}")
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def generate_test_scores(num_scores=100, mean=75, std_dev=15):
    """
    Generate and visualize normally distributed test scores between 0 and 100.
    
    Parameters:
    num_scores (int): Number of scores to generate (default: 100)
    mean (float): Mean score of the distribution (default: 75)
    std_dev (float): Standard deviation of the distribution (default: 15)
    
    Returns:
    list: List of test scores rounded to 1 decimal place
    """
    # Generate normally distributed scores
    scores = np.random.normal(mean, std_dev, num_scores)
    
    # Clip scores to ensure they're between 0 and 100
    scores = np.clip(scores, 0, 100)
    
    # Round to 1 decimal place and convert to list
    scores = [round(score, 1) for score in scores]
    
    # Create visualization
    plt.figure(figsize=(10, 6))
    
    # Create histogram
    plt.hist(scores, bins=20, density=True, alpha=0.7, color='skyblue', 
             label='Score Distribution')
    
    # Add a normal distribution curve
    x = np.linspace(0, 100, 100)
    norm_curve = stats.norm.pdf(x, mean, std_dev)
    plt.plot(x, norm_curve, 'r-', lw=2, label='Normal Curve')
    
    # Customize the plot
    plt.title('Distribution of Test Scores')
    plt.xlabel('Score')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Show the plot
    plt.show()
    
    return scores

# Example usage
if __name__ == "__main__":
    # Generate 100 test scores and visualize
    test_scores = generate_test_scores()
    
    # Print basic statistics
    print("Generated Test Scores:")
    print(test_scores)
    print(f"\nNumber of scores: {len(test_scores)}")
    print(f"Average score: {sum(test_scores)/len(test_scores):.1f}")
    print(f"Minimum score: {min(test_scores)}")
    print(f"Maximum score: {max(test_scores)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate uniform distribution of test scores (1000 scores between 0 and 100)
num_scores = 1000
test_scores = np.random.uniform(0, 100, num_scores)

# Plot the histogram to visualize the distribution
plt.figure(figsize=(10, 6))
plt.hist(test_scores, bins=20, edgecolor='black', alpha=0.7)
plt.title('Uniform Distribution of Test Scores', fontsize=14)
plt.xlabel('Test Scores', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate 1000 test scores with uniform distribution between 0 and 100
num_scores = 1000
test_scores = np.random.uniform(low=0, high=100, size=num_scores)

# Create the visualization
plt.figure(figsize=(10, 6))
plt.hist(test_scores, bins=20, edgecolor='black')

# Add labels and title
plt.xlabel('Test Scores')
plt.ylabel('Frequency')
plt.title('Uniform Distribution of Test Scores (0-100)')

# Add grid for better readability
plt.grid(True, alpha=0.3)

# Show the plot
plt.show()

# Print basic statistics
print(f"Number of scores: {len(test_scores)}")
print(f"Mean score: {np.mean(test_scores):.2f}")
print(f"Minimum score: {np.min(test_scores):.2f}")
print(f"Maximum score: {np.max(test_scores):.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(400)

# Generate 100 normally distributed prices
# Mean = 7.5 (midpoint of 5 and 10), SD = 1
# Then clip to ensure values stay between 5 and 10
num_prices = 100
mean_price = 7.50  # Midpoint of 5 and 10
std_dev = 1.0     # Standard deviation

prices = np.random.normal(mean_price, std_dev, num_prices)
prices = np.clip(prices, 5.00, 10.00)  # Ensure prices stay within 5-10 range
prices = np.round(prices, 2)           # Round to 2 decimal places

# Create the visualization
plt.figure(figsize=(10, 6))
plt.hist(prices, bins=15, edgecolor='black', color='skyblue')

# Add labels and title
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.title('Distribution of Product Prices ($5 - $10)')

# Add grid for better readability
plt.grid(True, alpha=0.3)

# Show the plot
plt.show()

# Print basic statistics
print(f"Number of prices: {len(prices)}")
print(f"Mean price: ${np.mean(prices):.2f}")
print(f"Minimum price: ${np.min(prices):.2f}")
print(f"Maximum price: ${np.max(prices):.2f}")
print(f"Standard deviation: ${np.std(prices):.2f}")
```
